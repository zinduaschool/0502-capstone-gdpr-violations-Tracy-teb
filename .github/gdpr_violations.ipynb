### Handling the missing values

```python
# Fill numeric columns with the mean or median
numeric_columns = gdpr_violations.select_dtypes(include=['float64', 'int64']).columns
gdpr_violations[numeric_columns] = gdpr_violations[numeric_columns].fillna(gdpr_violations[numeric_columns].median())
```

```python
### Filter by EU countries
eu_violations = gdpr_violations[gdpr_violations["Country"].str.contains("EU", case=False)]
```

### Time breakdown of the total violations in the EU

```python

eu_violations["date"] = pd.to_datetime(eu_violations["Date"])
eu_violations_by_year = (
    eu_violations.resample("Y", on="date")["Country"].count().reset_index(name="count")
)

plt.figure(figsize=(10, 6))
plt.bar(eu_violations_by_year["date"], eu_violations_by_year["count"], color="skyblue")
plt.xlabel("Year")
plt.ylabel("Number of Violations")
plt.title("Time Breakdown of Total GDPR Violations in the EU")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
try:
  eu_violations["date"] = pd.to_datetime(eu_violations["Date"])
except:
  print("Error converting 'Date' column to datetime. Check data format!")

# Check for empty DataFrame after resampling
eu_violations_by_year = (
    eu_violations.resample("Y", on="date")["Country"].count().reset_index(name="count")
)

if eu_violations_by_year.empty:
  print("eu_violations_by_year is empty. Check for missing data or EU violations!")
else:
  plt.figure(figsize=(10, 6))
  plt.bar(eu_violations_by_year["date"], eu_violations_by_year["count"], color="skyblue")
  plt.xlabel("Year")
  plt.ylabel("Number of Violations")
  plt.title("Time Breakdown of Total GDPR Violations in the EU")
  plt.xticks(rotation=45)
  plt.tight_layout()
  plt.show()
```
